## Z-Score

A Z-score gives you an idea of how far from the mean a data point is. But, more technically it's a measure of how many standard deviations below or above the population mean a raw score is. 

```python
def calculateZscore(z, population):
    score = (z - population.mean())/population.std()
    return score
```

## Standard error

Standard error is the sample mean error with the population mean error.

```python
def stdError(no_of_sample, population, sampleSize):
    std = sampleSTD(population, sampleSize)
    sample = {}
    for i in range(no_of_sample):
        sample[i] = np.random.choice(population, sampleSize).mean()
    error = std/np.sqrt(no_of_sample)
    
    return sample, error
```

```python
_, error = stdError(1000, marks, 4)
```

```python
def confidenceInterval(zScore, stdError, mean):
    upperLimit = mean + (zScore*stdError) 
    lowerLimit = mean - (zScore*stdError) 
    
    return lowerLimit, upperLimit
    
```

## Hypothesis Testing

We try to evaluate two mutual exclusive statement on a population using a sample of data.

**Steps**
1. Initial Assumption 
    - Null Hypothesis $H_0$
    - Alternative Hypothesis $H_a$
2. Collect evidence i.e. data
3. Analyse the data and make decision

||  $H_0$|$H_a$|
|-----|------|------|
|**Accept**|OK| Type 2 error |
|**Reject**|Type 1 error|OK  |


$H_0$ = Most students will score 60 or less than 60

$H_a$ = MOst students will score more than 60



-----------
- P-value => significance value - This helps us to reject or accept the $H_0$
    - if the p_value is less than 0.05 or 5% than we reject $H_0$ and accept the $H_a$
    - else we accept it and reject it respectively

```python
from scipy.stats import stats, chi2_contingency

import seaborn as sns
import pandas as pd
import numpy as np
dataset=sns.load_dataset('tips')
```

```python
dataset.head()
```

```python
dataset_table=pd.crosstab(dataset['sex'],dataset['smoker'])
print(dataset_table)
```

```python
#Observed Values
Observed_Values = dataset_table.values 
print("Observed Values :-\n",Observed_Values)
```

```python
val=chi2_contingency(dataset_table)
```

```python
val
```

```python
Expected_Values=val[3]
Expected_Values
```

```python
no_of_rows=len(dataset_table.iloc[0:2,0])
no_of_columns=len(dataset_table.iloc[0,0:2])
ddof=(no_of_rows-1)*(no_of_columns-1)
print("Degree of Freedom:-",ddof)
alpha = 0.05
```

```python
from scipy.stats import chi2
chi_square=sum([(o-e)**2./e for o,e in zip(Observed_Values,Expected_Values)])
chi_square_statistic=chi_square[0]+chi_square[1]
```

```python
print("chi-square statistic:-",chi_square_statistic)
```

```python
critical_value=chi2.ppf(q=1-alpha,df=ddof)
print('critical_value:',critical_value)
```

```python
#p-value
p_value=1-chi2.cdf(x=chi_square_statistic,df=ddof)
print('p-value:',p_value)
print('Significance level: ',alpha)
print('Degree of Freedom: ',ddof)
print('p-value:',p_value)
```

```python
if chi_square_statistic>=critical_value:
    print("Reject H0,There is a relationship between 2 categorical variables")
else:
    print("Retain H0,There is no relationship between 2 categorical variables")
    
if p_value<=alpha:
    print("Reject H0,There is a relationship between 2 categorical variables")
else:
    print("Retain H0,There is no relationship between 2 categorical variables")
```
